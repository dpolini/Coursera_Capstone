```python
import pandas as pd
import numpy as np
```

pd.read_html retrieves all the table of the webpage and creates a dataframe where the elements of the dataframe are dataframes contructed from the retrieved tables

```python
dfs=pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
```

We look at the result

```python
dfs
```

We see that the first table is the one of our interest and we retrieve it with the following command

```python
Toro=dfs[0]
Toro
```

We check the possible values in the column 'Boroughs' and then we drop the rows where no values is assigned

```python
Toro['Borough'].value_counts()
```

```python
Toro=Toro[Toro['Borough']!='Not assigned']
Toro=Toro.reset_index()
Toro
```

We check it worked

```python
Toro['Borough'].value_counts()
```

We rename the Neighbourhoods with a 'Not assigned' value to the value of the 'Borough'

```python
for i in range(0,len(Toro)):
    if Toro.iloc[i]['Neighbourhood']=='Not assigned':
        Toro.replace(Toro.iloc[i]['Neighbourhood'],Toro.iloc[i]['Borough'],inplace=True)   
```

We now group the Neighbourhood sharing the same postcode

```python
Toro_new=Toro.groupby(['Postcode','Borough'])['Neighbourhood'].apply(', '.join)
```

```python
Torof=Toro_new.to_frame().reset_index()
Torof
```
